## Reading in data from csv files

Here we import the csv package instead of using the numpy loadtxt routine

```python
#required packages and modules
import numpy as np
import matplotlib.pyplot as plt
import csv 
```

```python
# Lists which we want to fill
citynames, populations=[], []
```

```python
# Open CSV file, scan line-by-line and populate lists
with open('simplemaps-worldcities-basic.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    
    #skip the first line, which contains the legend
    next(readCSV)
    
    #add elements row-wise
    for row in readCSV:
        city = row[1]
        pop = int(float((row[4])))    # rounding of populations with .5s)
        country = row[5]
        
        if(country == 'China'):
            citynames.append(city)
            populations.append(pop)        
```

```python
# Sort lists by population (largest first) -- note we have converted to numpy arrays
sorting = np.argsort(populations)[::-1]
citysort, popsort = np.array(citynames)[sorting], np.array(populations)[sorting]
```

```python
for city, pop in zip(citysort, popsort):
    print ('%-18s %8i' % (city, pop) )
```
